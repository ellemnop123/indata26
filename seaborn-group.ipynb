```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
spotify = pd.read_csv("spotify-2023.csv", encoding ="latin1")
spotify.head()
```

```python
spotify.keys()
```

```python
hist1 = sns.histplot(data=spotify,x='bpm', kde = 'true')
hist1.plot()

#make sure to do the x.plot() bc VSCODE CANNOT ALWAYS HELP YOU :)
#kde = kernel density estimator
```

```python
rmgroup = spotify.groupby('released_month').count()['track_name']
# everything is case sensitive
rmgroup
```

```python
rmvalues = rmgroup.values
rmvalues
```

```python
rmindex = rmgroup.index

rmindex
```

```python
#Catplot next! scatterplot
cat1 = sns.catplot(x=rmindex, y=rmvalues)
plt.show()
```

```python
#now bar chart
cat1 = sns.catplot(x=rmindex, y=rmvalues, kind='bar')
plt.show()
```

```python
#now colors
cat1 = sns.catplot(x=rmindex, y=rmvalues, kind='bar', hue = rmindex) #saying index says what the hue is based on
#this is bad bc its treating the data as continuous when its categories
plt.show()
```

```python
#now we're adding color palette and legend (which you can edit and turn off and on)
cat1 = sns.catplot(x=rmindex, y=rmvalues, kind='bar', hue = rmvalues, palette = 'Spectral', legend = False)
#all this is case sensitive 

plt.show()
```

```python
rm_names = ["Jan", "Feb", "March", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
sns.color_palette("Spectral", 100) #tells it how many spectral colors to show
```

```python
cat4 = sns.catplot(x=rm_names, y=rmvalues, kind='bar', hue=rm_names, palette=cp, legend=False)
plt.show()
```

```python
spotify['month_cat'] = spotify['released_month'].astype('category')

spotify['month_cat'] = spotify['month_cat'].cat.rename_categories(rm_names)
```

```python
rel1 = sns.relplot(data=spotify, x="energy_%", y="bpm", hue='month_cat', palette=cp, style='mode', markers=['H', '>'],
 size = 'month_cat')

plt.show()
```
