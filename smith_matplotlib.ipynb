```python
cces = pd.read_csv('cces_example_data.csv')
cces.head()
```

```python
cces['sex_c'] = cces['sex'].astype('category')
cces['pid3_c'] = cces['pid3'].astype('category')
cces['white_c'] = cces['nonwhite'].astype('category')
cces['eval_c'] = cces['Evaluation'].astype('category')
```

## #5

```python

eval_groups = cces.groupby('eval_c').count()['Evaluation']
eval_groups = eval_groups.rename(index={0: 'Poor', 1: 'Fair', 2: 'Good', 3: 'Excellent'})

eval_groups
#official documentation i used (this took forever and I regret it but i did learn something atleast)
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html
```

```python
fig1, axs1 = plt.subplots()
col = ['#c2410a', '#e07f16', '#fcba03', '#6b9c2c']

axs1.bar(x = eval_groups.index, height = eval_groups.values, color=col)
plt.title('Evaluation Results')
# i wanted it to look nice/i hated the blue, so i figured out color
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

plt.show()
```
